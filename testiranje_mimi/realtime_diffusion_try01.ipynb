# Ahead of time audio processing

## High level idea

1. Realtime audio is piped into the system.
2. Streaming audio codec generates chunks and **Lowers the bitrate significantly**
3. Rolling buffer of size $B_{hist}$ gets populated with historic music data in real time
4. Previously generated future audio buffer of size $B_{future}$ is read
5. Rolling and future buffer get concatinated to form $B$, shich is fed to the model
6. Model diffuses the future buffer by one step
   1. Further most cuhnk is pure noise
   2. closest most chunk is 100% diffused predicted next chunk
